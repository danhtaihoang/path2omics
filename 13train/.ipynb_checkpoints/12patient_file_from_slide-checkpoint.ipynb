```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os,sys,time
from utils import *
```

```python
project = "BRCA"
n_cols_meta = 8

result_dir = f"{project}_results/"
df_actual = pd.read_pickle(f"{result_dir}{project}_actual.pkl")
df_actual
```

```python
## 
column_map1 = {col: "first" for col in df_actual.columns[:n_cols_meta]}
column_map2 = {col: "mean" for col in df_actual.columns[n_cols_meta:]}
column_map = {**column_map1, **column_map2}
column_map
```

```python
df_actual_patient = df_actual.groupby(["tissue_submitter_id"], as_index=False).aggregate(column_map)
df_actual_patient
```

```python
#df_pred = pd.read_csv(f"analysis_results/{project}_log_test_pred.csv")
df_pred = pd.read_pickle(f"{result_dir}{project}_pred.pkl")
df_pred
```

```python
df_pred_patient = df_pred.groupby(["tissue_submitter_id"], as_index=False).aggregate(column_map)
df_pred_patient
```

```python
actual_patient = df_actual_patient.iloc[:,n_cols_meta:].values
pred_patient = df_pred_patient.iloc[:,n_cols_meta:].values
print(actual_patient.shape, pred_patient.shape)
```

```python
coef, slope, padj = compute_coef_slope_padj(actual_patient, pred_patient)
print(sum(coef > 0.4),sum(coef > 0.45),sum(coef > 0.5), sum(coef > 0.55), sum(coef > 0.6))
```

```python
df_actual_patient.to_pickle(f"{result_dir}{project}_actual_patient.pkl")
df_pred_patient.to_pickle(f"{result_dir}{project}_pred_patient.pkl")

print(" --- completed creating files--- ")
```
